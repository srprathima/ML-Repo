```python
from sklearn.datasets import make_classification,make_s_curve,make_circles
import seaborn as sns
```

## Linearly Separable 

```python
X,y = make_classification(n_features=2,n_redundant=0,class_sep=3)
sns.scatterplot(x  =X[:,0],y = X[:,1],hue = y)
```

```python
from sklearn.svm import SVC
from mlxtend.plotting import plot_decision_regions
```

```python
svc = SVC(kernel='linear')
svc.fit(X,y)
plot_decision_regions(X,y,svc)
```

```python
X,y = make_classification(n_features=2,n_redundant=0,class_sep=0.4)
sns.scatterplot(x  =X[:,0],y = X[:,1],hue = y)
svc = SVC(kernel='linear')
svc.fit(X,y)
plot_decision_regions(X,y,svc)
```

## Non - LInear Data

```python
X,y = make_circles(n_samples=2000)
```

```python
sns.scatterplot(x  =X[:,0],y = X[:,1],hue = y)
```

```python
svc = SVC(kernel='linear')
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=23)
svc.fit(X_train,y_train)
y_pred = svc.predict(X_test)
print("Accuracy score:",accuracy_score(y_test,y_pred))
svc.fit(X,y)
plot_decision_regions(X,y,svc)
```

```python
svc = SVC(kernel='rbf')
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=23)
svc.fit(X_train,y_train)
y_pred = svc.predict(X_test)
print("Accuracy score:",accuracy_score(y_test,y_pred))
svc.fit(X,y)
plot_decision_regions(X,y,svc)
```
