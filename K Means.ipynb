```python
from sklearn.datasets import make_blobs
import seaborn as sns
```

```python
X,y = make_blobs(n_samples=60,random_state=23,cluster_std=3)
```

```python
sns.scatterplot(x = X[:,0],y = X[:,1])
```

```python
from sklearn.cluster import KMeans
```

```python
kmean = KMeans(n_clusters=3,init='random',random_state=42)
```

```python
kmean.fit(X)
```

```python
kmean.labels_
```

```python
centroids = kmean.cluster_centers_
centroids
```

```python
sns.scatterplot(x = X[:,0],y = X[:,1],hue = kmean.labels_,palette='tab10')
sns.scatterplot(x = centroids[:,0],y = centroids[:,1],s = 100,color = 'purple')
```
